Import necessary libraries and load the dataset.

```python
# Load the GDP dataset
df = pd.read_csv('gdp.csv')
print("‚úÖ Dataset loaded successfully!")
print(f"üìÅ File contains {df.shape[0]} rows and {df.shape[1]} columns")
```

## 3Ô∏è‚É£ Data Quality Assessment

Check for missing values, duplicates, and data quality issues.

```python
# Check for missing values
print("\n‚ùì Missing Values Analysis:")
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Percentage': missing_pct
})

missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)
print(f"\nColumns with missing values: {len(missing_df)}")
print(f"Total missing values: {missing.sum()}")
missing_df.head(10)
```

```python
# Visualize missing data
plt.figure(figsize=(14, 8))
sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis')
plt.title('Missing Data Heatmap (Yellow = Missing)', fontsize=16)
plt.xlabel('Columns (Years)', fontsize=12)
plt.ylabel('Countries/Regions', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Check for duplicates
print("\nüîÑ Duplicate Rows Check:")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Check for duplicate country names
duplicate_countries = df['Country Name'].duplicated().sum()
print(f"Duplicate country names: {duplicate_countries}")
```

```python
# Check unique values in key columns
print("\nüåç Unique Countries/Regions:")
print(f"Total unique countries/regions: {df['Country Name'].nunique()}")
print(f"Total unique codes: {df['Code'].nunique()}")
```

## 4Ô∏è‚É£ Data Preprocessing

Prepare data for analysis by handling data types and restructuring.

```python
# Create a copy for processing
df_clean = df.copy()

# Drop the 'Unnamed' column if it exists
if 'Unnamed: 65' in df_clean.columns:
    df_clean = df_clean.drop('Unnamed: 65', axis=1)
    print("‚úÖ Dropped 'Unnamed: 65' column")

print(f"\nüìä Cleaned dataset shape: {df_clean.shape}")
```

```python
# Convert year columns to numeric (they might be strings)
year_columns = [col for col in df_clean.columns if col not in ['Country Name', 'Code']]

print(f"\nüìÖ Converting {len(year_columns)} year columns to numeric...")
for col in year_columns:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

print("‚úÖ Conversion complete!")
```

```python
# Create a long-format dataframe for easier time-series analysis
df_long = df_clean.melt(
    id_vars=['Country Name', 'Code'],
    var_name='Year',
    value_name='GDP'
)

# Convert Year to integer
df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce')

# Remove rows with missing GDP values
df_long = df_long.dropna(subset=['GDP', 'Year'])

print("\nüìä Long-format dataset created:")
print(f"   Shape: {df_long.shape}")
df_long.head()
```

## 5Ô∏è‚É£ Statistical Summary

Analyze statistical properties of the GDP data.

```python
# Overall GDP statistics
print("\nüìà GDP Statistical Summary:")
df_long['GDP'].describe()
```

```python
# Statistics by decade
df_long['Decade'] = (df_long['Year'] // 10) * 10

print("\nüìä GDP Statistics by Decade:")
decade_stats = df_long.groupby('Decade')['GDP'].agg([
    'count', 'mean', 'median', 'std', 'min', 'max'
])
decade_stats
```

```python
# Top 10 countries by average GDP
print("\nüèÜ Top 10 Countries by Average GDP (1960-2020):")
top_countries = df_long.groupby('Country Name')['GDP'].mean().sort_values(ascending=False).head(10)
print(top_countries)
```

```python
# Bottom 10 countries by average GDP
print("\nüìâ Bottom 10 Countries by Average GDP (1960-2020):")
bottom_countries = df_long.groupby('Country Name')['GDP'].mean().sort_values(ascending=True).head(10)
print(bottom_countries)
```

## 6Ô∏è‚É£ Univariate Analysis

Analyze individual variables through distributions and visualizations.

```python
# GDP Distribution
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histogram
axes[0].hist(df_long['GDP'], bins=50, color='skyblue', edgecolor='black')
axes[0].set_xlabel('GDP ($)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].set_title('GDP Distribution (Histogram)', fontsize=14)
axes[0].grid(True, alpha=0.3)

# Box plot
axes[1].boxplot(df_long['GDP'].dropna(), vert=True)
axes[1].set_ylabel('GDP ($)', fontsize=12)
axes[1].set_title('GDP Distribution (Box Plot)', fontsize=14)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("üí° The distribution is highly skewed - a few countries have very high GDP values.")
```

```python
# Log-transformed GDP distribution (to handle skewness)
df_long['GDP_Log'] = np.log10(df_long['GDP'] + 1)  # +1 to avoid log(0)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histogram of log GDP
axes[0].hist(df_long['GDP_Log'], bins=50, color='coral', edgecolor='black')
axes[0].set_xlabel('Log10(GDP)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)
axes[0].set_title('Log-Transformed GDP Distribution', fontsize=14)
axes[0].grid(True, alpha=0.3)

# Density plot
df_long['GDP_Log'].plot(kind='density', ax=axes[1], color='coral', linewidth=2)
axes[1].set_xlabel('Log10(GDP)', fontsize=12)
axes[1].set_ylabel('Density', fontsize=12)
axes[1].set_title('Log-Transformed GDP Density', fontsize=14)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Number of countries with data per year
countries_per_year = df_long.groupby('Year')['Country Name'].count()

plt.figure(figsize=(14, 6))
plt.plot(countries_per_year.index, countries_per_year.values, marker='o', linewidth=2, color='green')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Countries', fontsize=12)
plt.title('Data Availability: Number of Countries Reporting GDP per Year', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nüìä Data coverage ranges from {countries_per_year.min()} to {countries_per_year.max()} countries per year")
```

## 7Ô∏è‚É£ Time Series Analysis

Examine how GDP has evolved over time globally and for specific countries.

```python
# Global GDP trend (sum of all countries per year)
global_gdp = df_long.groupby('Year')['GDP'].sum()

plt.figure(figsize=(14, 6))
plt.plot(global_gdp.index, global_gdp.values, linewidth=2.5, color='darkblue', marker='o', markersize=4)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Global GDP ($)', fontsize=12)
plt.title('Global GDP Trend (1960-2020)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("üí° Global GDP shows exponential growth, especially from the 1990s onwards.")
```

```python
# Average GDP per country per year
avg_gdp_per_year = df_long.groupby('Year')['GDP'].mean()

plt.figure(figsize=(14, 6))
plt.plot(avg_gdp_per_year.index, avg_gdp_per_year.values, linewidth=2.5, color='purple', marker='s', markersize=4)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average GDP per Country ($)', fontsize=12)
plt.title('Average GDP per Country Over Time (1960-2020)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# GDP trends for top 5 economies (by average GDP)
top_5_countries = df_long.groupby('Country Name')['GDP'].mean().sort_values(ascending=False).head(5).index

plt.figure(figsize=(14, 7))
for country in top_5_countries:
    country_data = df_long[df_long['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GDP'], marker='o', linewidth=2, label=country)

plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP ($)', fontsize=12)
plt.title('GDP Trends: Top 5 Economies (by Average GDP)', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# GDP growth rate calculation (year-over-year percentage change)
df_long_sorted = df_long.sort_values(['Country Name', 'Year'])
df_long_sorted['GDP_Growth_Rate'] = df_long_sorted.groupby('Country Name')['GDP'].pct_change() * 100

# Average growth rate per year globally
avg_growth = df_long_sorted.groupby('Year')['GDP_Growth_Rate'].mean()

plt.figure(figsize=(14, 6))
plt.plot(avg_growth.index, avg_growth.values, linewidth=2, color='red', marker='o', markersize=4)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average GDP Growth Rate (%)', fontsize=12)
plt.title('Average Global GDP Growth Rate (Year-over-Year %)', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("üí° Notice negative growth rates during global recessions (e.g., 2008-2009, 2020).")
```

## 8Ô∏è‚É£ Comparative Analysis

Compare GDP across different regions and time periods.

```python
# GDP by decade - Box plot
plt.figure(figsize=(14, 6))
df_long.boxplot(column='GDP', by='Decade', ax=plt.gca())
plt.xlabel('Decade', fontsize=12)
plt.ylabel('GDP ($)', fontsize=12)
plt.title('GDP Distribution by Decade', fontsize=14, fontweight='bold')
plt.suptitle('')  # Remove automatic title
plt.tight_layout()
plt.show()
```

```python
# Compare specific countries (example: BRICS nations)
brics_countries = ['Brazil', 'Russian Federation', 'India', 'China', 'South Africa']
df_brics = df_long[df_long['Country Name'].isin(brics_countries)]

plt.figure(figsize=(14, 7))
for country in brics_countries:
    country_data = df_brics[df_brics['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GDP'], marker='o', linewidth=2, label=country)

plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP ($)', fontsize=12)
plt.title('GDP Trends: BRICS Nations', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Top 15 countries in 2020 (most recent year)
df_2020 = df_long[df_long['Year'] == 2020].sort_values('GDP', ascending=False).head(15)

plt.figure(figsize=(14, 7))
plt.barh(df_2020['Country Name'], df_2020['GDP'], color='teal')
plt.xlabel('GDP ($)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Top 15 Countries by GDP in 2020', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()  # Highest at top
plt.tight_layout()
plt.show()
```

## 9Ô∏è‚É£ Correlation & Relationship Analysis

Analyze relationships between years and identify patterns.

```python
# Select a subset of years for correlation matrix (every 10 years)
year_subset = ['1960', '1970', '1980', '1990', '2000', '2010', '2020']
df_corr = df_clean[['Country Name'] + year_subset].set_index('Country Name')

# Calculate correlation matrix
corr_matrix = df_corr.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix: GDP Across Decades', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("üí° High correlation between consecutive decades indicates consistent economic patterns.")
```

```python
# Scatter plot: GDP in 1960 vs 2020
df_comparison = df_clean[['Country Name', '1960', '2020']].dropna()

plt.figure(figsize=(12, 8))
plt.scatter(df_comparison['1960'], df_comparison['2020'], alpha=0.6, s=100, color='darkgreen')
plt.xlabel('GDP in 1960 ($)', fontsize=12)
plt.ylabel('GDP in 2020 ($)', fontsize=12)
plt.title('GDP Comparison: 1960 vs 2020', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("üí° This shows how countries that had higher GDP in 1960 tend to have higher GDP in 2020.")
```

## üîü Advanced Insights

Derive deeper insights about economic development patterns.

```python
# Calculate GDP growth from 1960 to 2020
df_growth = df_clean[['Country Name', '1960', '2020']].dropna()
df_growth['Total_Growth'] = df_growth['2020'] - df_growth['1960']
df_growth['Growth_Percentage'] = ((df_growth['2020'] - df_growth['1960']) / df_growth['1960']) * 100

# Top 10 fastest growing economies (percentage)
top_growth = df_growth.sort_values('Growth_Percentage', ascending=False).head(10)

print("\nüöÄ Top 10 Fastest Growing Economies (1960-2020):")
print(top_growth[['Country Name', 'Growth_Percentage']])
```

```python
# Visualize top growing economies
plt.figure(figsize=(14, 7))
plt.barh(top_growth['Country Name'], top_growth['Growth_Percentage'], color='orange')
plt.xlabel('GDP Growth Percentage (1960-2020)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Top 10 Fastest Growing Economies by Percentage (1960-2020)', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Identify countries with complete data (no missing years)
year_cols = [col for col in df_clean.columns if col not in ['Country Name', 'Code']]
df_clean['Missing_Count'] = df_clean[year_cols].isnull().sum(axis=1)
complete_data_countries = df_clean[df_clean['Missing_Count'] == 0]['Country Name'].tolist()

print(f"\n‚úÖ Countries with complete data (1960-2020): {len(complete_data_countries)}")
print("\nExamples:")
print(complete_data_countries[:10])
```

```python
# Volatility analysis - standard deviation of GDP over time
df_volatility = df_long.groupby('Country Name')['GDP'].agg(['mean', 'std']).reset_index()
df_volatility['Coefficient_of_Variation'] = (df_volatility['std'] / df_volatility['mean']) * 100
df_volatility = df_volatility.sort_values('Coefficient_of_Variation', ascending=False)

print("\nüìä Top 10 Most Volatile Economies (by Coefficient of Variation):")
print(df_volatility.head(10)[['Country Name', 'Coefficient_of_Variation']])
```
